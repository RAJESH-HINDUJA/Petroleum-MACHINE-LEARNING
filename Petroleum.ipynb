To begin this exploratory analysis, first import libraries and define functions for plotting the data using matplotlib. Depending on the data, not all plots will be made.

```python
import pandas as pd   # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np   # linear algebra
import matplotlib.pyplot as plt   # plotting
import seaborn as sns  #visualization
import warnings
warnings.filterwarnings("ignore") 
```

Explore India's Growth in terms of Petroleum Products

```python
data = pd.read_csv("Petroleum Import and Export India Y-o-Y 1998 till 2023.csv")
data
```

To print the information of the data we can use data.info() command.

```python
data.info()
```

```python
# Changing from Long Format to Wide Format for Better Analysis
data = pd.melt(data,id_vars=['IMPORT/EXPORT','PRODUCT'],var_name='Fiscal Year',value_name='Volume')
data.info()
```

```python
data
```

Let’s see the mean, count , minimum and maximum values of the data

```python
data.describe()
```

```python
# Exploring PRODUCT type
data.PRODUCT.value_counts()
```

```python
# Remove the commas in the volume columns
data['Volume'] = data['Volume'].astype(str).str.replace(',','',regex=True)
```

The Volume is shown as a string. We need to change it to float values. Also, the strings are comma separated. So we will have to first replace the commas.

```python
# Change the Volume column from string to Float
data['Volume'] = data['Volume'].astype(float)
```

```python
data_crude = data[data['PRODUCT']=='CRUDE OIL']
```

```python
# Bar Plot for Crude Oil
plt.figure(figsize=(8,4))
sns.barplot(data_crude,x='Fiscal Year',y='Volume')
plt.xticks(rotation=90);
```

Analysis of Exported Petroleum Products from India

```python
# Filtering Exported Petroleum Products
data_products_export = data[(data['PRODUCT']!='CRUDE OIL') & (data['IMPORT/EXPORT']=='EXPORT')]
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data_products_export,x='Fiscal Year',y='Volume',hue='PRODUCT',size='PRODUCT')
plt.xticks(rotation=90);
```

```python
sns.catplot(data_products_export,x='Volume',y='Fiscal Year',hue='PRODUCT',kind='bar')
```

```python
plt.figure(figsize=(12,6))
fig = sns.relplot(data_products_export,x='Fiscal Year',y='Volume',col='PRODUCT', kind='line',col_wrap = 4,hue='PRODUCT',palette="husl")
fig.set_xticklabels()
```

Its evident from the above analysis export of finished or refined petroleum products from India has increased manifolds specially mainstream products like MS(Motor Spirit or Petrol) and HSD (High Speed Diesel).

HSD export growth is exponential

```python
# Filtering Imported Petroleum Products
data_products_import = data[(data['PRODUCT']!='CRUDE OIL') & (data['IMPORT/EXPORT']=='IMPORT')]
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data_products_import,x='Fiscal Year',y='Volume',hue='PRODUCT')
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(12,6))
fig = sns.relplot(data_products_import,x='Fiscal Year',y='Volume',col='PRODUCT', kind='line',col_wrap = 4,hue='PRODUCT',palette="husl")
fig.set_xticklabels()
```
